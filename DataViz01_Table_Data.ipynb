```python
import pandas as pd #ใช้สำหระบข้อมูลที่เป็นกราฟ
import os #ใช้สำหรับการนำเข้าไฟล์
```

```python
from google.colab import drive #กำหนดให้นำเข้าไฟล์จากไดร์ฟมาที่ google colab
drive.mount('/content/drive')
path = '/content/drive/My Drive/dataviz_2022_data'
```

```python
data_file_path = os.path.join(path,'confirmed-cases.xlsx')   
data_ori = pd.read_excel(data_file_path)   # อ่านไฟล์ excel
```

```python
data_ori #นำข้อมูลมาจาก drive
```

```python
data = data_ori[['sex','age','risk']] #เลือกเอาเฉพาะข้อมูลที่ต้องการใช้
data
```

```python
data.isnull().any() #เชคว่าภายในตารางมีค่าว่างหรือไม่
```

```python
data_risk1 = data.dropna() #ใช้ dropna เพื่อต้องการลบข้อมูลทั้งแถวออกเมื่อมีคอลัมน์ใดคอลัมน์หนึ่งเป็น NaN
data_risk1
```

```python
data_risk1.isnull().any()
```

```python
removes = data.shape[0] - data_risk1.shape[0]
print(f'size before drop = {data.shape[0]}')
print(f'size after drop = {data_risk1.shape[0]}')
print(f'we lost {100*(removes/data.shape[0])}% of data')
```

```python
Age = data[['age']].mean() #กำหนดให้ Age เป็นค่าเฉลี่ยของอายุ
data[['age']] = data[['age']].fillna(Age) #แทนค่าเฉลี่ยลงในข้อมูลที่ว่าง หรือเป็นค่า missing
```

```python
data[['risk']] = data[['risk']].fillna('ไม่ได้ระบุ') #แทนค่า Missing ด้วยคำว่า ไม่ได้ระบุ
```

```python
data[['sex']] = data[['sex']].fillna('ไม่ได้ระบุ') #แทนค่า Missing ด้วยคำว่า ไม่ได้ระบุ
```

```python
data.isnull().any()
```

```python
data['risk'].unique() #เลือกดูชื่อrisk
```

```python
risk_1 = data[data['risk']=='Sandbox'] #เลือก Sandbox
risk_1
```

```python
risk_2 = data[data['risk']=='State Quarantine'] #เลือก State Quarantine
risk_2
```

```python
risk_3 = data[data['risk']=='Samui plus'] #เลือกเอาเฉพาะ Samui plus
risk_3
```

```python
data_U = pd.concat([risk_1,risk_2,risk_3]) #รวมข้อมูลโดยใช้ .concat
data_U
```

```python
data_FM = data_U.replace('หญิง',1) #กำหนดเพศหญิงเป็น1
data_FM
```

```python
data_M = data_FM.replace('ชาย',150) #กำหนดเพศชายเป็น150
data_M
```

```python
data_N = data_M.replace('ไม่ได้ระบุ',300) #กำหนดไม่ระบุเพศเป็น300
data_N
```

```python
data_risk2 = data_N[['age','risk']] #เลือกเอาเฉพาะข้อมูลที่จะทำกราฟ
data_risk2
```

```python
data_N['risk'].unique() #เลือกดูชื่อrisk
```

```python
pd.plotting.parallel_coordinates(data_N,'risk',color=['k', '#FF0060', '#9F00FF']); #plot กราฟ
plt.xlabel('age')
plt.ylabel('sex,FM=1 M=150 other=300')
plt.title('Risk between sex and age')
plt.legend() #legend() คือ คำสั่งที่ใช้แสดงสัญลักษณ์ว่าเส้นไหนบอกค่าอะไร
plt.show()
```

สรุปผลได้ว่า เมื่อเปรียบเทียบเพศเเละอายุของผู้ป่วยจะเห็นได้ว่า ส่วนใหญ่ผู้ที่ติดเชื้อจะเป็นเพศหญิง เเละคนที่อายุมาก นอกจากนี้จะเห็นได้ว่าพื้นที่เสี่ยงมากที่มากที่สุดในสามพื้นที่ทำการเลือกมา คือ State Quarantine รองลงมา คือ samui plus และน้อยสุด คือ sanbox
